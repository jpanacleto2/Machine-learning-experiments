# LSTM_bicycle_Forecast

```python
import pandas as pd
import seaborn as sns
import matplotlib as mpl
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
from tensorflow.keras.layers import LSTM
```

```python
bike = pd.read_csv('bicycles.csv')
```

```python
bike.head()
```

```python
bike['datas'] = pd.to_datetime(bike['datas'])
```

```python
sns.lineplot(x='datas',y='contagem', data=bike)
plt.xticks(rotation=70)
```

# Scaling data

```python
sc = StandardScaler()
```

```python
sc.fit(bike['contagem'].values.reshape(-1,1))
```

```python
bike['contagem']
```

```python
y = sc.transform(bike['contagem'].values.reshape(-1,1))
```

```python
y
```

## Spliting int train and test

```python
train_size = int(len(bike)*0.9) #selecting 90% data to train
test_size = len(bike)-train_size #selecting 10% data to test
```

```python
Y_train = y[0:train_size]
Y_test = y[train_size:len(bike)]
```

```python
sns.lineplot(x='datas',y=Y_train[:,0],data=bike[0:train_size],label='train') 
sns.lineplot(x='datas',y=Y_test[:,0], data=bike[train_size:len(bike)],label='test')
plt.xticks(rotation=70) 
```

```python
vector = pd.DataFrame(Y_train)[0]
```

```python
vector
```

```python
def split_data(vector, n_steps):
    X_new, Y_new = [], []

    for i in range(n_steps, len(vector)):
        X_new.append(list(vector[i - n_steps:i]))
        Y_new.append(vector[i])


    X_new, Y_new = np.array(X_new), np.array(Y_new)

    return X_new, Y_new
```

```python
X_train_new, Y_train_new = split_data(vector,10)
```

```python
X_train_new[0:5]
```

```python
Y_train_new[0:5]
```

```python
vector2 = pd.DataFrame(Y_test)[0]
```

```python
X_test_new, Y_test_new = split_data(vector2,10)
```

# LSTM

Reshaping data for LSTM architecture

```python
X_train_new = X_train_new.reshape((X_train_new.shape[0],X_train_new.shape[1],1))
```

```python
X_train_new.shape
```

```python
X_test_new = X_test_new.reshape((X_test_new.shape[0],X_test_new.shape[1],1))
```

```python
X_test_new.shape
```

# Using LSTM

```python
recorrent = Sequential()
```

```python
recorrent.add(LSTM(128, input_shape=(X_train_new.shape[1],X_train_new.shape[2])
                    ))
recorrent.add(Dense(units=1))
```

```python
recorrent.compile(loss='mean_squared_error',optimizer='RMSProp')
recorrent.summary()
```

```python
result = recorrent.fit(X_train_new,Y_train_new,validation_data=(X_test_new,Y_test_new),epochs=100)
```

```python
y_adjusted = recorrent.predict(X_train_new)
```

```python
sns.lineplot(x='datas',y=Y_train[:,0],data=bike[0:train_size],label='train')
sns.lineplot(x='datas',y=y_adjusted[:,0],data=bike[0:15662],label='adjusted_train')
plt.xticks(rotation=70) 
```

```python
Y_pred = recorrent.predict(X_test_new)
```

```python
sns.lineplot(x='datas',y=Y_test[:,0], data=bike[train_size:len(bike)],label='test')
sns.lineplot(x='datas',y=Y_pred[:,0], data=bike[train_size+10:len(bike)],marker='.',label='prediction')
plt.xticks(rotation=70) 
```
