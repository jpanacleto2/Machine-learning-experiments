This project uses NeuralNetworks to classify tumors as benign or Malignant using the Scikit Learn Breast cancer dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow import keras
```

# Loading de data

```python
# loading dataset
dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(dataset)
```

```python
# loading the data to a data frame
data_frame = pd.DataFrame(dataset.data, columns = dataset.feature_names)
```

```python
data_frame
```

```python
data_frame['class'] = dataset.target
```

```python
data_frame.head()
```

```python
data_frame.shape
```

```python
# checking the distribution of Target Varibale
data_frame['class'].value_counts()
```

1 --> Benign

0 --> Malignant

# Separating the features and target

```python
X = data_frame.drop(columns='class', axis=1)
Y = data_frame['class']
```

```python
Y
```

# Splitting the data into training data & Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Data padronization

```python
X_train
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

```python
X_train_std[0]
```

# Building Neural Network

![](https://miro.medium.com/v2/resize:fit:1400/0*4ohJSbscb4MxOzAI)

```python
tf.random.set_seed(3)

# setting up the layers of Neural Network

model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),
                          keras.layers.Dense(30, activation='relu'),
                          keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# compiling the Neural Network

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the Meural Network

history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

# Ploting accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

# Acuraccy of the model on test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print('loss: ',loss,'accuracy: ',accuracy)
```

```python
print(X_test_std.shape)
print(X_test_std[0])
```

model.predict() gives the prediction probability of each class for that data point

```python
Y_pred = model.predict(X_test_std)
```

```python
print(Y_pred.shape)
print(Y_pred[0])
```

```python
# converting the prediction probability to class labels

Y_pred_labels = [np.argmax(i) for i in Y_pred]
print(Y_pred_labels)
```

**Building the predictive system**

```python
input_data = (11.76,21.6,74.72,427.9,0.08637,0.04966,0.01657,0.01115,0.1495,0.05888,0.4062,1.21,2.635,28.47,0.005857,0.009758,0.01168,0.007445,0.02406,0.001769,12.98,25.72,82.98,516.5,0.1085,0.08615,0.05523,0.03715,0.2433,0.06563)

# change the input_data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array as we are predicting for one data point
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardizing the input data
input_data_std = scaler.transform(input_data_reshaped)

prediction = model.predict(input_data_std)
print(prediction)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if(prediction_label[0] == 0):
  print('The tumor is Malignant')

else:
  print('The tumor is Benign')
```
